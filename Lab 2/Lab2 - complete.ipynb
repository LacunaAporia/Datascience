```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print(sns.get_dataset_names())

MtG_df = []
MtG_df = pd.read_csv('https://raw.githubusercontent.com/mtgjson/mtgcsv/master/csv/ISD.csv', low_memory=False)

flight_df = sns.load_dataset('flights')
health_df = sns.load_dataset('healthexp')
exercise_df = sns.load_dataset('exercise')
geyser_df = sns.load_dataset('geyser')
sea_df = sns.load_dataset('seaice')
titan_df = sns.load_dataset('titanic')


ans_df = sns.load_dataset('anscombe')

```

```python
print('\nflights\n')
 # The goal is to chart the number of flights each year by a span of months. 
 # This chart demonstrates a colaboration of two styles The line plot with averaged wings, 
 # and the simulated lined scatter plot as it conveys a lot of information in a very human readable format. 
print(flight_df)
# flight_df.groupby('year').plot(kind='scatter', x = 'month', y = 'passengers')
sns.lineplot(x='year',y='passengers',data=flight_df) 
plt.title('Flights') 
sns.lineplot(x='year',y='passengers',hue='month',data=flight_df, marker='o', linestyle=' ')
print('\n')
```

```python

print('\nGeyser Survey\n') 
# It's hard to get away from the scatter graph it is so informative 
# and frequently feels like the primium option for conveying varied data. 
print(geyser_df)
sns.scatterplot(data=geyser_df, y = 'waiting', x = 'duration', hue='kind')

# guis_df.groupby('kind').plot(kind = 'scatter', y = 'waiting', x = 'duration')
plt.title(label='Geyser Survey') 
print('\n')
```

```python
print('\nSea Ice Loss\n')
print(sea_df)
sea_df.plot(kind = 'hist', x = 'Date', y = 'Extent')
# Not sure what's up with the histogram plotting with seaborn, but 
# sns.histplot(data=sea_df, x = 'Date')
plt.title('Sea Ice') 
print('\n')
```

```python
print('\nHealthExp\n')
print(health_df)
#health_df.plot(kind = 'scatter', y = 'Year', x = 'Life_Expectancy')
#sns.barplot(x='Year',y='Life_Expectancy',hue='Country',data=health_df,ci=0)
sns.scatterplot(data=health_df,x='Year', y = 'Life_Expectancy',hue='Country')
plt.title('Health - Expense Analysis') 
print('\n')
```

```python
print('\nHealthExp\n') 
# False reporting, skewing the original dataset to only show certian variables
#  and therby truely showing the numbers but misrepresenting their meaning.
# Design is also ugly and introduces chart clutter without adding content. 
print(health_df)
# sns.histplot(data=health_df, y = 'Life_Expectancy', x = 'Year', hue='Country', multiple='stack')
sns.histplot(data=health_df,  y = 'Life_Expectancy', x = 'Year', hue = 'Country') 
sns.scatterplot(data=health_df,  y = 'Life_Expectancy', x = 'Year', hue = 'Country') 

# dsns.histplot(data=health_df, y = 'Spending_USD', x = 'Year', hue = 'Country')
plt.title(label='Life Expectancy Since 1970 per Country') 
print('\n')
```

```python
print('\nExercise\n')
# I chose the Swarmplot style as it gives highly detailed info and can be split amnogst catagories. 
print(exercise_df)
# exer_df.groupby('kind')
# sns.lineplot(data=exercise_df, x='time', y='pulse', hue = 'diet')
sns.swarmplot(data=exercise_df, y = 'time', x = 'pulse', hue = 'kind')
sns.catplot(data=exercise_df, y = 'time', x = 'pulse', kind='boxen', col = 'kind', legend=False)
# sns.rugplot(data=exercise_df, x='time', y='pulse', hue='diet')
# plt.title(label='Exercise') 
print('\n')
```

```python
print('Anscombe\n')
# I like this set because it demonstrates common data annomalies
print(ans_df)
# - hexbin style, added nothing, made the graph less asthetically pleasing
ans_df.groupby('dataset').plot(kind = 'scatter', x = 'x', y= 'y', style = 'dataset')
plt.title(label='Anscombe') 
print('\n')
```

```python
print('Ticket Costs Aboard the Titanic\n')
print(titan_df)
sns.catplot(data=titan_df, x='class', y='fare', kind='boxen', col = 'embark_town', legend=False)
print()
sns.rugplot(data=titan_df, x='class', y='fare', hue='sex', height=.02)
print('\n')
```
