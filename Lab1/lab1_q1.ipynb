```python
def data_url(city_name):
    base_url = "https://facultyweb.cs.wwu.edu/~wehrwes/courses/csci141_21s/fp/data/"
    return base_url + city_name + ".csv"

tSum = 0.0                      #  Values for holding the leading city in precipitation
tAvg = 0.0                      #
tCity = "AZ_LosFrancisco"       #

import pandas as pd
cities = ["AZ_Phoenix", "CA_LosAngeles", "CA_SanFrancisco"] # "CO_Denver" , "FL_Jacksonville", "GA_Atlanta", "IL_Chicago", "MA_Boston", "MN_MinneapolisStPaul", "NY_Ithaca", "NY_NewYorkJFK", "OR_Portland", "PA_Philadelphia", "TX_DallasFtWorth", "UT_SaltLake",  "VA_WashingtonDulles", "WA_Bellingham", "WA_Seattle"]

for city in cities:
 df = pd.read_csv(data_url(city), low_memory=False)
 
 df["DailyPrecipitation"]                          
 mask = df["REPORT_TYPE"].str.contains("SOD")  # Masking to isolate the start of day readings. 
 dailies = df[mask]
 dailies
 
 precip_amts = dailies["DailyPrecipitation"].astype(str).str.replace("T", "0.0") # Sanitizing data to be more process friendly
 precip_amts = precip_amts.str.rstrip("s").astype(float)                         #
  
 print(city)
 print(" DailyPrecipitation ")
 print(" Anual Total AVG: ")
 print(precip_amts.mean())
 if precip_amts.mean() >= tAvg:
  tAvg = precip_amts.mean()
  tCity = city


 print(" Anual Total SUM:")
 print(precip_amts.sum())
 if precip_amts.sum() >= tSum:
  tSum = precip_amts.sum()
  tCity = city
 print("\n")

 
 print(" Standard Deviation:")
 print(precip_amts.std()) # standard deviation
 precip_amts.plot(kind = 'box')
 precip_amts.plot()
 print(precip_amts.describe()) # show a bunch of useful summary data
 print('\n\n')

print("The city with the highest precipitation by average was " + tCity)
```
